```python
from openpyxl import Workbook
from openpyxl import load_workbook
from os import listdir
import pandas as pd
```

```python
datasheets = listdir("eacad_data")
```

```python
for name in datasheets:
    name_no_ext = name.split(".")[0]
    wb = load_workbook(f"eacad_data/{name}")
    ws = wb[wb.sheetnames[0]]
    cols_to_delete = [3,5,7,10,12,13,16,18,21]
    i = 0
    for col in cols_to_delete:
        ws.delete_cols(col-i)
        i += 1
    ws.delete_rows(1,4)
    df = pd.DataFrame(ws.values)
    df.columns = df.iloc[0]
    df = df.drop([0])
    df = df.astype({'S.No': 'int32', "Vacancy": "int32", "Current Strength": "int32"})
    df = df.set_index(["S.No"])
    df['Instructor'] = df['Instructor'].str.strip()
    df.to_csv(f"acad_data_clean/{name_no_ext}.csv")
    print(f"Cleaned up {name_no_ext}")
```
